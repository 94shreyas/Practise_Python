```python
sns.lineplot(
    x='Year',
    y='NA_Sales',
    data=ice_hockey_data
)
plt.xlabel("NA_Sales")
plt.ylabel("Year")
plt.title("Sales Trend for Ice_Hockey")
plt.show()
```

```python
#Setting Limits on X or Y axis
#plt.xlim
#plt.ylim

sns.lineplot(
    x='Year',
    y='NA_Sales',
    data=ice_hockey_data
)
plt.xlim(2000,2020)
plt.xlabel("NA_Sales")
plt.ylabel("Year")
plt.title("Sales Trend for Ice_Hockey")
plt.show()
```

```python
baseball_data=data.loc[data['Name']=='Baseball']
```

```python
#Distribution of Genre among top 3
#Top 3 data = Top 3 Publisher, Top 3 genre and Top 3 platform#

top3_publishers=data['Publisher'].value_counts().index[:3]
top3_platforms=data['Platform'].value_counts().index[:3]
top3_genres=data['Genre'].value_counts().index[:3]
```

```python
top3_publishers
```

```python
top3_data = data.loc[
     (data["Publisher"].isin(top3_publishers))
     & (data["Platform"].isin(top3_platforms))
     & (data["Genre"].isin(top3_genres))
     ]
```

```python
sns.countplot(
    x='Publisher',
    hue='Platform',
    data=top3_data
)
```

```python
#metrics for sales in every genre

sns.boxplot(
    y='Global_Sales',
    x='Publisher',
    data=top3_data
)
```

```python
#metrics for sales in every genre

sns.barplot(
    y='Global_Sales',
    x='Publisher',
    data=top3_data
)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(1,3,1)
sns.scatterplot(
    x=top3_data['NA_Sales'],
    y=top3_data['EU_Sales'],
)
plt.subplot(1,3,2)
sns.scatterplot(
    x=top3_data['NA_Sales'],
    y=top3_data['JP_Sales'],
)
plt.subplot(1,3,3)
sns.scatterplot(
    x=top3_data['NA_Sales'],
    y=top3_data['Other_Sales'],
)
```
