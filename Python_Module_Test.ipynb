```python
import numpy as np

# Create an array of numbers from 1 to 99
original_array = np.arange(1, 100)

# Reshape the array into 3 rows and 33 columns
reshaped_array = original_array.reshape(3, 33)

# Split the reshaped array vertically into three equal parts
split_arrays = np.vsplit(reshaped_array, 3)

# Calculate the sum of each part
sums_of_parts = [np.sum(split_array) for split_array in split_arrays]

print("Original Array:")
print(original_array)

print("\nReshaped Array:")
print(reshaped_array)

print("\nSum of each part:")
print(sums_of_parts)
```

```python
import numpy as np

# Create an array of integers from 0 to 9
original_array = np.arange(10)

# Separate even and odd numbers
even_numbers = original_array[original_array % 2 == 0]
odd_numbers = original_array[original_array % 2 != 0]

# Stack even numbers before odd numbers horizontally
new_array = np.hstack((even_numbers, odd_numbers))

# Reshape the new array to have 2 rows and 5 columns
reshaped_array = new_array.reshape(2, 5)

# Find the mean of the numbers in the first row
mean_first_row = np.mean(reshaped_array[0])

# Display the results
print("Original Array:")
print(original_array)

print("\nEven Numbers:", even_numbers)
print("Odd Numbers:", odd_numbers)

print("\nNew Array (even numbers before odd numbers):")
print(new_array)

print("\nReshaped Array (2 rows, 5 columns):")
print(reshaped_array)

print("\nMean of numbers in the first row:", mean_first_row)
```

```python
import numpy as np

# Create an array of integers from 1 to 10
original_array = np.arange(1, 11)

# Extract even and odd numbers
even_numbers = original_array[original_array % 2 == 0]
odd_numbers = original_array[original_array % 2 != 0]

# Stack the two separated arrays vertically
resultant_array = np.vstack((even_numbers, odd_numbers))

# Display the results
print("Original Array:", original_array)
print("Even Numbers:", even_numbers)
print("Odd Numbers:", odd_numbers)
print("\nResultant Array (Stacked vertically):")
print(resultant_array)
print("\nShape of the Resultant Array:", resultant_array.shape)
```

```python
import numpy as np

# Create an array of integers from 1 to 20
original_array = np.arange(1, 21)

# Square the even numbers and leave the odd numbers unchanged
squared_array = np.where(original_array % 2 == 0, original_array**2, original_array)

# Reshape the array into 10 rows and 2 columns
reshaped_array = squared_array.reshape(10, 2)

# Find the sum of the last column
sum_last_column = np.sum(reshaped_array[:, 1])

# Display the results
print("Original Array:", original_array)
print("Squared Array (even numbers squared, odd numbers unchanged):", squared_array)
print("\nReshaped Array (10 rows, 2 columns):")
print(reshaped_array)
print("\nSum of the last column:", sum_last_column)
```

```python
import numpy as np

# Create an array of numbers from 1 to 10
original_array = np.arange(1, 11)

# Extract even and odd numbers
even_numbers = original_array[original_array % 2 == 0]
odd_numbers = original_array[original_array % 2 != 0]

# Stack the two separated arrays vertically
stacked_array = np.vstack((even_numbers, odd_numbers))

# Display the results
print("Original Array:", original_array)
print("Even Numbers:", even_numbers)
print("Odd Numbers:", odd_numbers)
print("\nStacked Array (Vertically):")
print(stacked_array)

# Calculate the mean of the last two columns
mean_last_two_columns = np.mean(stacked_array[:, -2:])

# Display the mean of the last two columns
print("\nMean of the last two columns:", mean_last_two_columns)
```
