```
#fillter

condition_1=(mckinsey['year']==2007)
condition_2=(mckinsey['continent']=='Americas')
condition_3=(mckinsey['life_exp']>75)

mask=(condition_1)&(condition_2)&(condition_3)
mask

df=mckinsey[mask]
```
